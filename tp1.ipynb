```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as  pd
```

```python
#df = pd.read_csv("_data.csv")
df = pd.read_csv("2020_ct.csv", delimiter=";")
pd.options.display.float_format = "{:.3f}".format
```

```python
def convert_to_numeric(_df, col):
    _df[col] = pd.to_numeric(_df[col])
    return _df
```

```python
def change_commas(_df, col):
    _df[col] = df[col].str.replace(".", "")
    _df[col] = df[col].str.replace(",", ".")
    return _df
```

##### Parse amounts

```python
change_commas(df, "Definitivo")
change_commas(df, "Vigente")
change_commas(df, "Devengado")
change_commas(df, "Sanción")

convert_to_numeric(df, "Definitivo")
convert_to_numeric(df, "Vigente")
convert_to_numeric(df, "Devengado")
convert_to_numeric(df, "Sanción")
```

```python
# Same code different description
def same_id_same_desc(_df, col_name):
    _df = _df[_df.groupby([col_name])[col_name+"_desc"].transform('nunique') > 1]
    #_df = _df[_df.duplicated([col_name, col_name+"_desc"])]
    return _df
```

```python
# Same description different code
def same_desc_same_id(_df, col_name):
    _df = _df[_df.groupby([col_name+"_desc"])[col_name].transform('nunique') > 1]
    return _df
```

```python
def draw_boxplot(_df, var, outliers=False):
    plt.boxplot(_df[var], showfliers=outliers)
    title = print("Boxplot variable '{}'".format(var))
    plt.title(title)
```

```python
def draw_distr(_df, var, maxy=0.25):
    fig, ax = plt.subplots(figsize=(15,30))
    proy = calculate_frequency(_df, var)
    ax.scatter(x=proy[var], y=proy["frequency"])
    ax.set_xticklabels(proy[var].str.upper(), fontdict={'horizontalalignment': 'right', 'size':12}, rotation=50)
    ax.vlines(x=proy[var], ymin=0, ymax=proy["frequency"], color='blue')
    ax.set_ylim(0, maxy)
    title ='Distribución de los porcentajes para los valores de la variable {}'.format(var) 
    plt.title(title)
    plt.show()
```

```python
def calculate_frequency(_df, col):
    return _df[col].value_counts(normalize=True).reset_index(name='frequency').sort_index(ascending=True).rename(columns={'index': col})
```

```python
df_biv_dev = df[["Definitivo", "Geo", "Geo_desc"]].copy(deep=True)
df_biv_dev = df_biv_dev.groupby(["Geo_desc"], as_index=False).sum()
df_biv_dev["Definitivo"] = df_biv_dev["Definitivo"] / 1000000
df_biv2["Total"] = df_biv_san["Sanción"] - df_biv_dev["Definitivo"]
```

```python

ax = df_biv_dev.Definitivo.plot(y='Definitivo', kind="bar", color='red', position=0)
ax = df_biv_san["Sanción"].plot(y='Sanción', kind="bar", color='green', position=1)

plt.rcParams["figure.figsize"] = (16,6)

plt.xticks([i for i in range(18)], df_biv_san["Geo_desc"])
plt.title("Gasto previsto (Sanción) vs gasto total (Definitivo) por comuna")
plt.ylabel("Monto en millones")
plt.legend(["Definitivo", "Sanción"])
plt.show()
```
